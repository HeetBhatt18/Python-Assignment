# **Q3. Find GCD using Euclidean Algorithm**

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

a = num1
b = num2

while b != 0:
  remainder = a % b
  print(f"{a} = {b} × {a // b} + {remainder}")
  a = b
  b = remainder

gcd = a

print(f"\nGCD of {num1} and {num2} is: {gcd}")
```
