 # **Q1. Check whether a number is prime**

```python
user_input = input("Enter a number to check if it's prime: ")
number = int(user_input)

if number < 2:
    print(f"{number} is not a prime number")
    exit()

if number == 2:
    print(f"{number} is a prime number")
    exit()

if number % 2 == 0:
    print(f"{number} is not a prime number")
    exit()

limit = int(number ** 0.5) + 1

is_prime_number = True

for i in range(3, limit, 2):
    if number % i == 0:
        is_prime_number = False
        break

if is_prime_number:
    print(f"{number} is a prime number!")
else:
    print(f"{number} is not a prime number.")
```

# **Q2. Compute sum of squares: 1^2 + 2^2 + ... + n^2.**

```python
user_input = input("Enter a number (n) to compute sum of squares up to n: ")
n = int(user_input)

print("\n--- Using Loop Method ---")
sum_of_squares = 0

for i in range(1, n + 1):
  square = i ** 2
  sum_of_squares = sum_of_squares + square

  print(f"{i}² = {square}, Running total: {sum_of_squares}")

print(f"\nSum of squares from 1² to {n}² = {sum_of_squares}")

print("\n--- Using Formula Method ---")

formula_result = n * (n + 1) * (2 * n + 1) // 6
print(f"Using formula: {n} × ({n}+1) × (2×{n}+1) ÷ 6 = {formula_result}")

print(f"\nBoth methods match: {sum_of_squares == formula_result}")
```

# **Q3. Find GCD using Euclidean Algorithm**

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

a = num1
b = num2

while b != 0:
  remainder = a % b
  print(f"{a} = {b} × {a // b} + {remainder}")
  a = b
  b = remainder

gcd = a

print(f"\nGCD of {num1} and {num2} is: {gcd}")
```

# **Q5. Reverse a number without string conversion.**

```python
num = int(input("Enter a number to reverse: "))

original_num = num

is_negative = False
if num < 0:
    is_negative = True
    num = abs(num)

reversed_num = 0

while num > 0:

    last_digit = num % 10

    reversed_num = reversed_num * 10 + last_digit

    num = num // 10

if is_negative:
    reversed_num = -reversed_num

print(f"Original number: {original_num}")
print(f"Reversed number: {reversed_num}")
```

# **Q6. Check Armstrong number.**

### An Armstrong number (also called narcissistic number) is a number that equals the sum of its own digits each raised to the power of the number of digits.

```python
num = int(input("Enter a number to check: "))

original_num = num

num_digits = 0
temp = num
while temp > 0:
    num_digits += 1
    temp = temp // 10

sum_of_powers = 0
temp = num

while temp > 0:

    digit = temp % 10

    power = digit ** num_digits

    sum_of_powers = sum_of_powers + power

    temp = temp // 10

if sum_of_powers == original_num:
    print(f"{original_num} is an Armstrong number!")
else:
    print(f"{original_num} is NOT an Armstrong number.")
```

# **Q8. Count digits, vowels, consonants in a string.**

```python
text = input("Enter a string: ")

digit_count = 0
vowel_count = 0
consonant_count = 0

vowels = "ariouAEIOU"

for char in text:
  if char.isdigit():
    digit_count += 1

  elif char.isalpha():
    if char in vowels:
            vowel_count += 1
    else:
      consonant_count += 1

print("\n--- Results ---")
print(f"Original string: {text}")
print(f"Number of digits: {digit_count}")
print(f"Number of vowels: {vowel_count}")
print(f"Number of consonants: {consonant_count}")
print(f"Total letters: {vowel_count + consonant_count}")
```

# **Q9. Find largest and second largest in a list.**

```python
numbers = input("Enter numbers separated by spaces: ").split()

for i in range(len(numbers)):
  numbers[i] = int(numbers[i])

if len(numbers) < 2:
  print("List should be atleast 2 numbers!")

else:
  largest = numbers[0]
  second_largest = numbers[0]

for num in numbers:
  if num > largest:
    largest = num

for num in numbers:
  if num!= largest:
    if second_largest == numbers[0] or num > second_largest:
      second_largest = num

print(f"List: {numbers}")
print(f"Largest number: {largest}")
print(f"Second largest number: {second_largest}")
```

# **Q10. Factorial using recursion.**

```python
def factorial(n):
  if n==0 or n==1:
    return 1

  else:
    return n * factorial(n-1)

num = int(input("Enter a number to find a factorial: "))

if num < 0:
  print("Factorial is not defined for negative numbers!")

else:
  result = factorial(num)
  print(f"{num}! = {result}")
```

 # **Q11. Check palindrome using function.**

 ### A palindrome is a word, phrase, or number that reads the same forwards and backwards.

```python
def is_palindrome(text):

  reversed_text = text[::-1]

  if text == reversed_text:
    return True
  else:
    return False

word = input("Enter a word or phrase: ")

if is_palindrome(word):
  print(f"{word} is a palindrome number")

else:
  print(f"{word} is not a palindrome number")
```

# **Q14. Nth Fibonacci using recursion.**

```python
def fibonacci(n):
  if n==0:
    return 0
  elif n==1:
    return 1
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter the position (n) to find Fibonacci number: "))

if n < 0:
    print("Please enter a non-negative number!")
else:
    result = fibonacci(n)
    print(f"Fibonacci({n}) = {result}")
```

# **Q15.  Frequency of elements in list.**

```python
def count_frequency(lst):

    frequency = {}

    for element in lst:
        if element in frequency:
            frequency[element] += 1
        else:
            frequency[element] = 1

    return frequency

my_list = [1, 2, 3, 2, 1, 3, 3, 4, 5, 2, 1]

result = count_frequency(my_list)

print("Original list:", my_list)
print("Frequency of elements:")
for element, count in result.items():
    print(f"{element} appears {count} times")
```

# **Q16. Check perfect number.**

```python
def is_perfect(num):

    if num <= 0:
        return False

    divisor_sum = 0

    for i in range(1, num):
        if num % i == 0:
            divisor_sum = divisor_sum + i

    if divisor_sum == num:
        return True
    else:
        return False

num = int(input("Enter a number to check: "))

if is_perfect(num):
    print(f"{num} is a perfect number!")
else:
    print(f"{num} is NOT a perfect number.")
```

# **Q17. Find primes in range.**

```python
def is_prime(num):

    if num < 2:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True

def find_primes_in_range(start, end):

    primes = []

    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)

    return primes

start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

prime_numbers = find_primes_in_range(start, end)

print(f"\nPrime numbers between {start} and {end}:")
if prime_numbers:
    print(prime_numbers)
    print(f"\nTotal prime numbers found: {len(prime_numbers)}")
else:
    print("No prime numbers found in this range.")
```
