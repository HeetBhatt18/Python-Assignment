# **Q8. Count digits, vowels, consonants in a string.**

```python
text = input("Enter a string: ")

digit_count = 0
vowel_count = 0
consonant_count = 0

vowels = "ariouAEIOU"

for char in text:
  if char.isdigit():
    digit_count += 1

  elif char.isalpha():
    if char in vowels:
            vowel_count += 1
    else:
      consonant_count += 1

print("\n--- Results ---")
print(f"Original string: {text}")
print(f"Number of digits: {digit_count}")
print(f"Number of vowels: {vowel_count}")
print(f"Number of consonants: {consonant_count}")
print(f"Total letters: {vowel_count + consonant_count}")
```
