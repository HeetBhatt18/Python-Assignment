# **Q7. Generate Fibonacci series.**



```python
n = int(input("Enter the number of terms: "))

a = 0
b = 1

if n < 0:
    print("Please enter a positive integer!")
elif n == 0:
  print("Fibonacci series up to", n, "term:")
  print(a)
else:
    print("Fibonacci series up to", n, "terms:")
    print(a ,b ,end =" ")

for i in range(2, n):
  c = a + b
  print(c, end=" ")

  a = b
  b = c
```
