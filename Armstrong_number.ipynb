# **Q6. Check Armstrong number.**

### An Armstrong number (also called narcissistic number) is a number that equals the sum of its own digits each raised to the power of the number of digits.

```python
num = int(input("Enter a number to check: "))

original_num = num

num_digits = 0
temp = num
while temp > 0:
    num_digits += 1
    temp = temp // 10

sum_of_powers = 0
temp = num

while temp > 0:

    digit = temp % 10

    power = digit ** num_digits

    sum_of_powers = sum_of_powers + power

    temp = temp // 10

if sum_of_powers == original_num:
    print(f"{original_num} is an Armstrong number!")
else:
    print(f"{original_num} is NOT an Armstrong number.")
```
