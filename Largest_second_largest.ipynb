# **Q9. Find largest and second largest in a list.**

```python
numbers = input("Enter numbers separated by spaces: ").split()

for i in range(len(numbers)):
  numbers[i] = int(numbers[i])

if len(numbers) < 2:
  print("List should be atleast 2 numbers!")

else:
  largest = numbers[0]
  second_largest = numbers[0]

for num in numbers:
  if num > largest:
    largest = num

for num in numbers:
  if num!= largest:
    if second_largest == numbers[0] or num > second_largest:
      second_largest = num

print(f"List: {numbers}")
print(f"Largest number: {largest}")
print(f"Second largest number: {second_largest}")
```
